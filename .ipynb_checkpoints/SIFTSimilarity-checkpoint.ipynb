```python
keypoints_1, descriptors_1 = pysift.computeKeypointsAndDescriptors(img1)
```

```python
keypoints_2, descriptors_2 = pysift.computeKeypointsAndDescriptors(img2)
```

```python
keypoints_3, descriptors_3 = pysift.computeKeypointsAndDescriptors(img3)
```

```python
keypoints_4, descriptors_4 = pysift.computeKeypointsAndDescriptors(img4)
```

```python
keypoints_5, descriptors_5 = pysift.computeKeypointsAndDescriptors(img5)
```

```python
#feature matching
bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
```

```python
matches = bf.match(descriptors_1,descriptors_2)
matches = sorted(matches, key = lambda x:x.distance)
```

```python
matches2 = bf.match(descriptors_3,descriptors_2)
matches2 = sorted(matches2, key = lambda x:x.distance)
```

```python
matches3 = bf.match(descriptors_4,descriptors_5)
matches3 = sorted(matches3, key = lambda x:x.distance)
```

```python
img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[:50], img2, flags=2)
plt.imshow(img3),plt.show()
```

```python
result_img2 = cv2.drawMatches(img3, keypoints_3, img2, keypoints_2, matches2[:50], img2, flags=2)
plt.imshow(result_img2),plt.show()
```

```python
result_img3 = cv2.drawMatches(img4, keypoints_4, img5, keypoints_5, matches3[:50], img5, flags=2)
plt.imshow(result_img3),plt.show()
```
